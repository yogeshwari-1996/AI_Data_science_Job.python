```python
import pandas as pd
from sqlalchemy import create_engine

# Load full dataset
df = pd.read_csv(r"C:\AI & Data Science Job - Sheet1 (1).csv")

print("Total rows:", len(df))   # should show 10345

# Connect to PostgreSQL
engine = create_engine("postgresql://postgres:Krishna%401996@localhost:5432/job_db")

# Upload
df.to_sql("jobs", engine, if_exists="replace", index=False)
```

```python
##Top Job Roles##

query = """
select job_title, count(*) as total_jobs
from jobs
group by job_title
order by total_jobs desc
limit 10;
"""

pd.read_sql(query,engine)
```

```python
##Average Salary by Role###

query = """
select job_title, Round(AVG(salary)::numeric,2) as avg_salary
from jobs
group by job_title
order by avg_salary desc;
"""

pd.read_sql(query,engine)
```

```python
##Top Hiring country##

import pandas as pd
from sqlalchemy import create_engine

# connect
engine = create_engine("postgresql://postgres:Krishna%401996@localhost:5432/job_db")

# query
query = """
SELECT country, COUNT(*) AS job_count
FROM jobs
GROUP BY country
ORDER BY job_count DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```

```python
## Most In demand skills##

query = """
SELECT skills_ml, COUNT(*) AS demand
FROM jobs
GROUP BY skills_ml
ORDER BY demand DESC
LIMIT 10;
"""

pd.read_sql(query, engine)
```

```python
## Data Visualization###

import matplotlib.pyplot as plt
import seaborn as sns

top_roles = pd.read_sql ("""
select job_title,count(*) as total
from jobs
group by job_title
order by total desc
limit 10;
""", engine)

sns.barplot(x="total", y="job_title", data=top_roles)
plt.title("Top Job Roles")
plt.show()
```
